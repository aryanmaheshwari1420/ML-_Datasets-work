## Data Preprocessing

```python
dataset = dataset.drop_duplicates()
print(dataset)
```

```python
dataset.isnull().head()
```

```python
print("Total Null-values in :" , 
      dataset.isnull().sum())
```

```python
dataset.dropna(how='any',subset=['Country','Purchased'],inplace=True) 
#inplace = True --> we are doing in the same dataset
#any --> removing that rows of country and purchased where NaN are available
```

```python
print(dataset)
```

```python
x = dataset[['Country','Age','Salary']].values
y = dataset['Purchased'].values
# print(x,y)
```

```python
from sklearn.impute import SimpleImputer
help(SimpleImputer)
print(x[:,1:3])
```

```python
imputer = SimpleImputer(missing_values=np.nan,strategy= 'mean')
imputer.fit(x[1:,1:3])
x[:,1:3] = imputer.transform(x[:,1:3])
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_test , y_train = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(y_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:, 4:] = sc.fit_transform(x_train[:,4:])
```
