```python
df = df.drop_duplicates()
df
```

```python
df.replace("?", np.nan, inplace = True)
df.head(100)
```

```python
print("Total Null-values in :\n" , df.isnull().sum())
```

```python
df.dropna(how='any',subset=['horsepower'],inplace=True) 
```

```python
x = df[['mpg','cylinders','horsepower','weight','year','acceleration','displacement']].values
y = df['origin'].values
```

```python
from sklearn.impute import SimpleImputer
help(SimpleImputer)
print(x[:,1:3])
```

```python
imputer = SimpleImputer(missing_values=np.nan,strategy= 'mean')
imputer.fit(x[1:,1:3])
x[:,1:3] = imputer.transform(x[:,1:3])
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_test , y_train = train_test_split(x,y,test_size=0.2,random_state=2)
```
